```python
# 2. Объединение данных
merged_df = pd.merge(df1, df2, left_on="Номер контракта", right_on="NumContract", how="left")

merged_df.head()
```

```python
# 3. Исключение клиентов по условиям
filtered_df = merged_df[
    ~merged_df["ContractID"].isin(cession_df["ContractID"])
    & ~merged_df["Status"].isin(["BANKRUPTCY", "REPAID", "EARLY_REPAID"])
]

filtered_df.head()
```

```python
# 4. Построение сводной таблицы по регионам
pivot_table = filtered_df.groupby("SubdivisionName").agg(
    Количество_клиентов=("Номер контракта", "count"),
    Сумма_выданных_займов=("Сумма выдачи", "sum"),
).reset_index()

print("Сводная таблица по регионам:")
print(pivot_table)
```

```python
# 5. Создание базы данных SQLite
conn = sqlite3.connect("loans.db")
filtered_df.to_sql("loans", conn, if_exists="replace", index=False)
```

```python
# 6. SQL-запросы
# Запрос 1: Клиенты с регионом "онлайн" и выданные в 2024 году
query1 = """
SELECT l.* FROM loans AS l
WHERE SubdivisionName = 'Онлайн' AND issueDateTimestamp LIKE '2024%'
"""
result1 = pd.read_sql(query1, conn)
print("Клиенты с регионом 'Онлайн' и выданные в 2024 году:")
print(result1)

```

```python
# Запрос 2: Количество клиентов и сумма займов по статусам
query2 = """
SELECT l.Status, COUNT(*) AS Количество_клиентов, SUM(LoanAmount) AS Сумма_выданных_займов
FROM loans AS l
GROUP BY Status
"""
result2 = pd.read_sql(query2, conn)
print("Количество клиентов и сумма выданных займов по статусам:")
print(result2)

```

```python
# Запрос 3: Все поля из отчетов 1 и 2
query3 = """
SELECT l.* FROM loans AS l
"""
result3 = pd.read_sql(query3, conn)
print("Все поля из отчетов 1 и 2:")
print(result3)

```
